```python
import pandas as pd

df = pd.read_csv('img_all_info.csv')
df
```

```python
city = "Kyoto"
test_df = pd.read_csv('1/test/0.2-attribute-'+city+'.filtered.txt',header=None,names=['photo_id','uid','tid','pid','attribute'])
test_df
```

```python
train_df = pd.read_csv('1/train/0.2-attribute-'+city+'.filtered.txt',header=None,names=['photo_id','uid','tid','pid','attribute'])
train_df
```

```python
all_df = pd.concat([test_df,train_df])
all_df
```

```python
from datetime import datetime

raw_photo_id = df['Photo/video identifier'].values.tolist()
raw_data_taken = df['Data taken'].values.tolist()

photo_id_list = all_df['photo_id'].values.tolist()
data_taken_list = []

for photo_id in photo_id_list:
    if photo_id in raw_photo_id:
        i = raw_photo_id.index(photo_id)
        timestamp = datetime.timestamp(datetime.strptime(raw_data_taken[i], '%Y-%m-%d %H:%M:%S.%f'))
        # timestamp = datetime.strptime(raw_data_taken[i],"%Y-%m-%d %H:%M:%S")
        data_taken_list.append(timestamp)
    else:
        print('no such photo_id')
    # break

all_df["data_taken"] = data_taken_list
all_df
```

```python
result = all_df.drop('photo_id',axis=1).drop('tid',axis=1).drop('attribute',axis=1)
result
```

```python
result.to_csv("checkins/"+city+"_checkins.txt",sep=' ',index=False,header=False)
```
