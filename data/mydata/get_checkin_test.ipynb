```python
import pandas as pd

city = "Vien"
persent = "0.2"
# df = pd.read_csv('img_all_info.csv')
df = pd.read_csv('all_place_info.csv')
df
```

```python
test_df = pd.read_csv('2/test/'+persent+'-attribute-'+city+'.filtered.txt',header=None,names=['photo_id','uid','tid','pid','attribute'])
test_df
```

```python
from datetime import datetime

# raw_photo_id = df['Photo/video identifier'].values.tolist()
raw_photo_id = df['photo_id'].values.tolist()
raw_data_taken = df['Data taken'].values.tolist()

photo_id_list = test_df['photo_id'].values.tolist()
data_taken_list = []

for photo_id in photo_id_list:
    if photo_id in raw_photo_id:
        i = raw_photo_id.index(photo_id)
        timestamp = datetime.timestamp(datetime.strptime(raw_data_taken[i], '%Y-%m-%d %H:%M:%S.%f'))
        # timestamp = datetime.strptime(raw_data_taken[i],"%Y-%m-%d %H:%M:%S")
        data_taken_list.append(timestamp)
    else:
        print('no such photo_id')
    # break

test_df["data_taken"] = data_taken_list
test_df
```

```python
result = test_df.drop('photo_id',axis=1).drop('tid',axis=1).drop('attribute',axis=1)
result
```

```python
result.to_csv("checkins/"+city+"_"+persent+"_checkins_test_2.txt",sep=' ',index=False,header=False)
```
